# WebScraping PY

o programa abre o site do Waze e imprime o código usando BeautifulSoup

```python
import urllib
from bs4 import BeautifulSoup
meulink="https://www.waze.com/pt-BR/livemap?zoom=17&lat=-8.12942&lon=-34.90322"
html_do_link = urllib.request.urlopen(meulink)
soup = BeautifulSoup(html_do_link)
print(soup)
```
